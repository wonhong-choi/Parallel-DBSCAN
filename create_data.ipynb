```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
def create_moon_data(n_samples, noise=0.1):
    noisy_moons = datasets.make_moons(n_samples=n_samples, noise=noise) # n_samples 데이터 생성 갯수, noise=노이즈 정도

    x,y=[],[]
    for data in noisy_moons[0]:
        x.append(round(data[0],3)*125*(n_samples//2000)) # *250은 scaling을 위함
        y.append(round(data[1],3)*125*(n_samples//2000)) # *250은 scaling을 위함
    plt.figure(figsize=(10,10))
    plt.scatter(x,y,s=10) # s= plot시 point의 size
    plt.title('noisy_moon', size=15) # set title
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    plt.show()
    
    # noisy moon data
    f=open("./Release/data/moon.txt","w")
    f.write(str(n_samples)+'\n')
    for i in range(len(x)):
        f.write(str(x[i])+' '+str(y[i])+'\n')
    f.close()

def create_circle_data(n_samples, noise=0.05):
    # n_samples 데이터 생성 갯수, noise=노이즈 정도
    noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=noise)

    x,y=[],[]
    for data in noisy_circles[0]:
        x.append(round(data[0],3)*375*(n_samples//4000)) # *750은 scaling을 위함
        y.append(round(data[1],3)*375*(n_samples//4000)) # *750은 scaling을 위함
    plt.figure(figsize=(10,10))
    plt.scatter(x,y,s=10)
    plt.title('noisy_circle', size=15) # set title
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    plt.show()

    # noisy circle data
    f=open("./Release/data/circle.txt","w")
    f.write(str(n_samples)+'\n')
    for i in range(len(x)):
        f.write(str(x[i])+' '+str(y[i])+'\n')
    f.close()
```

```python
if __name__=='__main__':
    print("필요 라이브러리 : numpy, matplotlib, sklearn !")
    n_samples=int(input('데이터 갯수 입력:')) # 데이터 개수 조절
    
    
    np.random.seed(0)
    
    create_moon_data(n_samples,0.1)
    create_circle_data(n_samples*2,0.06)
```
