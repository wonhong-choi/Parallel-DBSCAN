```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_data(FILENAME):
    x=[]
    y=[]
    cluster=[]

    f=open(FILENAME,"r")

    n_datas = int(f.readline())
    
    print('total:', n_datas) # 전체 데이터 갯수 출력

    for line in f.readlines():
        line=line.split(' ')
        #print(line)
        x.append(float(line[0]))
        y.append(float(line[1]))
        cluster.append(int(line[2]))

    f.close()

    x=np.array(x)
    y=np.array(y)
    cluster=np.array(cluster)

    fig, ax = plt.subplots(figsize=(10,10)) ## 전체 figure size 지정 (width,height)
    for g in np.unique(cluster):
        ix = np.where(cluster == g)
        ax.scatter(x[ix], y[ix], label = g, s = 15) # s== plot시 point 크기
    ax.legend()
    plt.title(FILENAME[15:], size=15) # set title
    plt.show()
```

```python
if __name__=='__main__':
    #print("출력하고자 하는 파일명을 입력하세요. ex) 'moon.txt'\n")
    
    FILENAME=input('"출력하고자 하는 파일명을 입력하세요. ex) \'moon.txt\'\n>>>') # 출력하고자 하는 file명 /// 이 부분만 수정하면 됨
    
    FILENAME0='./Release/data/'+FILENAME[:-4]+'_serial.txt'
    FILENAME1='./Release/data/'+FILENAME[:-4]+'_parallel1.txt'
    FILENAME2='./Release/data/'+FILENAME[:-4]+'_parallel2.txt'
    plot_data(FILENAME0)
    plot_data(FILENAME1)
    plot_data(FILENAME2)
```
